```python
import spacy 
from bs4 import BeautifulSoup
```

```python
nlpSites = {
    '1': 'https://www.techtarget.com/searchenterpriseai/definition/natural-language-processing-NLP',
    '2': 'https://www.ibm.com/cloud/learn/natural-language-processing',
    '3': 'https://www.sas.com/en_us/insights/analytics/what-is-natural-language-processing-nlp.html',
    '4': 'https://hbr.org/2022/04/the-power-of-natural-language-processing',
    '5': 'https://cloud.google.com/learn/what-is-natural-language-processing'
}
```

```python
HTML_TAGS = [
    '<br>',
    '<strong>',
    '</strong>',
    '<b>',
    '</b>',
    '\xa0',
    '\xad',
    '\n',
    '--',
    ' _ ',
    '\t'
]
```

```python
for site in nlpSites:
  !wget '{nlpSites[site]}' -O '{site}'
```

```python
def words_to_string(list_of_words: list):
  str = ''
  for word in list_of_words:
    if list == type(word):
      word = words_to_string(word)
    if '' != word:
      str += '{} '.format(word)
  return str
```

```python
def format_str(str: str):
  for HTMLTags in HTML_TAGS:
    str = str.replace(HTMLTags, '')
  list_of_words = str.split(' ')
  str = words_to_string(list_of_words)
  return str
```

```python
get_sentences = lambda str: str.split(',')
```

```python
def to_1d(lista: list):
  new_list = []
  for obj in lista:
    if list == type(obj):
      new_list += to_1d(obj)
    else:
      new_list.append(obj)
  return new_list
```

```python
corpus = {}
for site in nlpSites:
  doc = open(f'./{site}', 'r')
  soup = BeautifulSoup(doc, 'html.parser')
  corpus[site] = to_1d([get_sentences(format_str(a.get_text())) for a in soup.find_all('p')])
```
